# Data Visualization
Data visualisation is the process fo representing data graphically to reveal patterns, trends, correlation and outliers.
## Use Case:
Use these tools to analyze datasets like sales records, scientific measurements

### Install/Import Matplotlib

```python
import matplotlib.pyplot as plt
```

### Basic Setup

```python
student = ['daniel', 'alfred', 'amah', 'divine', 'ruth', 'george']
scores = [80, 90, 70, 60, 85, 74]
```

#### Line Plot

```python
plt.plot(student, scores, color = 'blue')

plt.title('Student Test Score')
plt.xlabel('Student Name')
plt.ylabel('Scores')
plt.show()
```

### Scatter Plot

```python
plt.scatter(student, scores, color='red')
plt.title('Student Test Score')
plt.xlabel('Student Name')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(student, scores, color='green')
plt.title('Student Scores')
plt.xlabel('Student Name')
plt.ylabel('Scores')
plt.show()
```

```python
plt.hist(scores, color='purple')
plt.title('Student Score')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4, 2))

plt.bar(student, scores, color='green')
plt.title('Student Scores')
plt.xlabel('Student Name')
plt.ylabel('Scores')
plt.show()
```

## Seaborn
Seaborn is a super easy tool in python that helps you draw pretty plots

#### Install/Import Seaborn

```python
import seaborn as sns
import pandas as pd
```

### Basic Setup

```python
student = ['daniel', 'alfred', 'amah', 'divine', 'ruth', 'george']
scores = [80, 90, 70, 60, 85, 74]

data = pd.DataFrame({'Student': student, 'Scores': scores})
print(data)
```

### Line plot

```python
sns.lineplot(x='Student', y='Scores', data=data, color='blue')
plt.title('Student Test Scores')
plt.show()
```

### Scatter Plot

```python
sns.scatterplot(x='Student', y='Scores', data=data, color='red')

plt.title('Student Test Scores')
plt.show()
```

```python
sns.barplot(x='Student', y='Scores', data=data, color='green')

plt.title('Student Test Score')
plt.show()
```

## Histogram

```python
sns.histplot(data=data, x='Scores', color='purple')
plt.title('How scores pile up')
plt.show()
```

### Visualizing Web3 Token Transfer Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\HP USER\Documents\Data Analyst\python-tutorial\Module_2\web3_token_transfers.csv', parse_dates=['timestamp'])
df['date'] = df['timestamp'].dt.date

# Set Seaborn theme
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df['amount'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Token Transfer Amounts')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.xlim(0, 3000)
plt.show()
```

### Daily data transfer

```python
daily_volume = df.groupby('date')['amount'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=daily_volume, x='date', y='amount', marker='o')
plt.title('Daily Total Volume')
plt.xlabel('Date')
plt.ylabel('Total Amount Transfered')
```
